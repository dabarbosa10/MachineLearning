## ML for physicist ##

The prediction depends on the available number of data points, noise and the relevant features.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_train=np.random.uniform(0,1,10)
```

```python
y_train=2*x_train
```

```python
from matplotlib import rcParams
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x_train,y_train,'.')
plt.grid()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import datasets, linear_model
```

```python
N_train=10
x=np.linspace(0.01,0.99,N_train)
sigma=0.5
#True Model
def truem(x,s):
    return 2*x+s*np.random.randn(len(x))
```

```python
plt.figure(figsize=(8, 6))
y1=truem(x,sigma)
plt.plot(x,y1,"o",ms=8,alpha=0.5,label='Data')
plt.grid()
```

```python
pol=PolynomialFeatures(degree=10)
X=pol.fit_transform(x[:,np.newaxis])
clf=linear_model.LinearRegression()
clf.fit(X,y1)
```

```python
xp=np.linspace(0.02,0.98,200)
xt=pol.fit_transform(xp[:,np.newaxis])
plt.figure(figsize=(8, 6))
plt.plot(x,y1,"o",ms=8,alpha=0.5,label='Data')
plt.plot(xp,clf.predict(xt))
plt.grid()
```
